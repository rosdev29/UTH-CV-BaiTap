## Tra cứu PDF — từng người

File: [`Computer_Vision 24.pdf`](./Computer_Vision%2024.pdf).

| Người | Mục PDF | Việc | Nhắc khi code |
|-------|---------|------|---------------|
| **Duy** | **§1.2–1.4.2** | Blur + Sobel | Grayscale trước; **blur trước, Sobel sau**; \(M=\sqrt{G_x^2+G_y^2}\); \(\theta=\arctan(G_y/G_x)\); **đừng** dùng \(\|G_x\|+\|G_y\|\) |
| **Tuyền** | **§1.4.3** | NMS | Lượng hoá θ về 0°/45°/90°/135°; so sánh M(p) với 2 hàng xóm dọc hướng gradient; θ **vuông góc** cạnh |
| **Trường** | **§1.4.4–1.6** | Hysteresis + Otsu + IoU | τ_high / τ_low; **DFS/BFS** 8-kết nối (đừng single-pass); Otsu trên histogram M → τ_high, τ_low = τ_high/2; so sánh `cv2.Canny` bằng **IoU** |
| **Nguyên** | **§2.1–2.3.1** | Hough multi-scale | Normal form \(\rho=x\cos\theta+y\sin\theta\); ≥2 mức (coarse/fine); coarse-to-fine: tìm ứng viên coarse rồi refine fine |
| **Nhung** | **§2.3.2–2.4** | Top-k + gradient voting | Top-k peaks + NMS trên accumulator; windowed voting theo θ gradient (±window°, cyclic trên [0°,180°)); thí nghiệm k=1,3,5 và F1 vs window |
| **Quyến** | **§3.2–3.7** | RANSAC | Generic: `fit_fn`, `distance_fn`; chọn **2 hình** s&lt;5 (vd. đường thẳng s=2, đường tròn s=3); giải thích ≥100 từ/hình; \(N=\log(1-p)/\log(1-(1-e)^s)\); outlier 20/40/60% |

**Làm chung (nhận xét / đồ thị):**
- **Trường + Tuyền:** Otsu vs ngưỡng tay; khi nào Otsu tốt/xấu (§1.5–1.6); điều chỉnh σ / ngưỡng khi cạnh giả.
- **Nguyên + Nhung:** So sánh top-k trên ảnh bàn cờ; đồ thị F1 vs window size (§2.4).
- **Quyến:** Viết giải thích chọn 2 hình dạng; báo cáo N lý thuyết vs thực nghiệm (§3.3–3.7).
